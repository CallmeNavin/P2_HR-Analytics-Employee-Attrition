```python
from google.colab import files
uploaded = files.upload()
import pandas as pd

df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.head()
df.shape
df.info()
df.isnull().sum()

df['Attrition'].value_counts(normalize=True)
df['Gender'].value_counts(normalize=True)
df['Age']

df['Attrition_encoded'] = df['Attrition'].map({'Yes': 1, 'No': 0})
df['Age_group'] = pd.cut(df['Age'], bins = [0, 30, 40, 100], labels = ['<30', '30 - 40', '>=40'], right = False)
df['Income_group'] = pd.qcut(df['MonthlyIncome'], q = 3, labels = ['Low', 'Medium', 'High'])

df.head()
df.to_csv('HR_clean.csv', index = False)

from google.colab import files
files.download('HR_clean.csv')
```
